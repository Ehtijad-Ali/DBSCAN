```python
'''Imigine you are working for a E-Commerce company and the marketing team wants
to target customers more efficitively by creating segments based on their
purchasing behavior.However traditional segmentation methods are not providing
staisfactory result
'''
```

```python
# DBSCAN : is used to identify clusters of the customers with similar purchasing patterns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
np.random.seed(42)
data = np.random.rand(100, 2)
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
# apply DBSCAN

dbscan  = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(data_scaled)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis', marker='o', s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
result_df = ({'Feature 1' : data[0],
              'Feature 2' :data[1], 'Clusters': clusters

})
```

```python

result_df = pd.DataFrame({'clusters': clusters})

print('Number of clusters:', len(np.unique(clusters)))
print('Size of each cluster:')
print(result_df['clusters'].value_counts())
```

```python
# model testing user input

user_input = np.array([[0.6, 0.7]])
user_input_scaled = scaler.transform(user_input)
user_cluster = dbscan.fit_predict(user_input_scaled)
print('user input belongs to cluster', user_cluster[0])
```
